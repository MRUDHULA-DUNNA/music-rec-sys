```python
df['lyrics'] = df['lyrics'].str.lower().replace(r'^\w\s',' ').replace(r'\n', ' ', regex = True)
```

```python
import nltk
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    stemming = [stemmer.stem(w) for w in tokens]
    return " ".join(stemming)
```

```python
df['text'] = df['lyrics'].apply(lambda x: tokenization(x))
```

```python
 from sklearn.feature_extraction.text import TfidfVectorizer
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['lyrics'])
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
df[df['track_name']=='goodbye']
```

```python
def recommendation(song_df):
    idx = df[df['track_name'] == song_df].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    songs = []
    for m_id in distances[1:10]:
        songs.append(df.iloc[m_id[0]].track_name)

    return songs
```

```python
recommendation('cry')
```
